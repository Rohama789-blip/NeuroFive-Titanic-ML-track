```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/train[1].csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df = df.drop(columns=['Cabin'])
```

```python
df.isnull().sum()
```

## Missing Values Handling

- The **Age** column was filled with the median because it is a numerical feature and the median is less affected by outliers.
- The **Embarked** column was filled with the mode because it is a categorical feature.
- The **Cabin** column was dropped because it contained a large number of missing values.

```python
plt.figure(figsize=(6,4))
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df['Fare'])
plt.title("Fare Boxplot")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
